# Exercise 1: Try Linear Regression just using numpy (Without Tensorflow/Pytorch or other torch library). 

```python
import numpy as np
```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70]], dtype='float32')
```

```python
# Target (apples)
targets = np.array([[56], 
                    [81], 
                    [119], 
                    [22], 
                    [103]], dtype='float32')
```

```python
print("Inputs:",inputs)
print("Targets:",targets)
```

```python
from numpy import random

w = random.rand(1, 3)
b=random.rand(1,1)
print(w)
print(b)
```

```python
# Define the model
# @=FOR DOT PRODUCT
def model(x):
    return x @ w.transpose()+b
```

```python
# Generate predictions
pred=model(inputs)
print(pred)
```

```python
# Compare with targets
print(targets)
```

## Loss Function

We can compare the predictions with the actual targets, using the following method: 
* Calculate the difference between the two matrices (`preds` and `targets`).
* Square all elements of the difference matrix to remove negative values.
* Calculate the average of the elements in the resulting matrix.

The result is a single number, known as the **mean squared error** (MSE).

```python
# MSE loss
def mse(t1,t2):
    diff=(t1-t2)**2
    _sum_=np.sum(diff)   
    return _sum_/diff.size
```

```python
# Compute loss
loss=mse(pred,targets)
print(loss)
l_rate=1e-4
```

## Gradient Descent Algorithm

```python
def grad_desc(w,b):
    m=len(targets)
    b_de=(l_rate * ((1/m) * np.sum(model(inputs) - targets)))
    for i in range(len(w)):
        w_de=(l_rate * ((1/m) * np.sum(model(inputs) - targets)))
    return w_de,b_de
```

## Training for 200 epochs

```python
# Train for 200 epochs
for i in range(200):
    pred=model(inputs)
    loss=mse(pred,targets)
    w_de,b_de=grad_desc(w,b)
    w=w-w_de
    b=b-b_de
```
